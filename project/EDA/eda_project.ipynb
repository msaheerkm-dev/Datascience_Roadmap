OUTLIER HANDLING
- Outliers are data points that deviate significantly from the rest of the observations.  
They can occur due to data entry errors, measurement issues, or genuine extreme behavior.

###  Detection Methods
- **Statistical Approaches:**  
  - **Z-Score Method:** Detects values lying beyond a threshold (commonly ±3).  
  - **IQR Method:** Identifies values below *(Q1 - 1.5 × IQR)* or above *(Q3 + 1.5 × IQR)*.  
- **Visualization Techniques:**  
  - Boxplots, scatter plots, and histograms help visually detect outliers.

###  Handling Techniques
- **Capping (Winsorization):** Replace extreme values with nearest valid limits.  
- **Transformation:** Apply log or square root transformation to reduce skewness.  
- **Deletion:** Remove outlier rows only if they are incorrect or irrelevant. 

```python
# visualising outliers 
plt.figure(figsize=(12,8))
sns.boxplot(df[df.select_dtypes('number').columns])
plt.xticks(rotation=45)
```

```python
# selecting feature with outliers
df1=['Age','CGPA']
df1
```

```python
# outlier handling  usign IQR method
outliers_df = pd.DataFrame(columns=[ 'Outlier_Value', 'Outlier_Feature'])

for i in df1:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    IQR=q3-q1
    lo=q1-1.5*IQR
    hi=q3+1.5*IQR

    outliers = df[(df[i] < lo) | (df[i] > hi)][[i]].copy()
    outliers['Outlier_Value'] = outliers[i]  # store outlier value
    outliers['Outlier_Feature'] = i
    outliers = outliers[[ 'Outlier_Value', 'Outlier_Feature']]

    # Append to final DataFrame
    if not outliers.empty:  
        outliers_df = pd.concat([outliers_df, outliers], ignore_index=True)

    # capping step
    df[i] = np.where(df[i] < lo, lo, np.where(df[i] > hi, hi, df[i]))
    # winzorise This simply replaces extreme outliers with boundary values — no math errors.


# Reset index for clarity
outliers_df.reset_index(drop=True, inplace=True)

print(outliers_df)


```

```python
# boxplot after outlier capping
plt.figure(figsize=(12,8))
sns.boxplot(df[df.select_dtypes('number').columns])
plt.xticks(rotation=35)
```

##  Univariate Analysis

**Univariate Analysis** focuses on analyzing a single feature (variable) at a time to understand its distribution and basic statistics.  
It helps identify patterns such as skewness, spread, and presence of outliers.

- **For numerical data:** Use histograms, boxplots, and summary statistics (mean, median, mode).  
- **For categorical data:** Use bar plots or count plots to view category frequency.

### 📦 Boxplot
A **boxplot** visually displays the distribution of a numerical variable and highlights potential outliers.  
It shows:
- **Median (Q2)**
- **Quartiles (Q1 and Q3)**
- **Interquartile Range (IQR = Q3 - Q1)**
- **Outliers** as individual points beyond whiskers.

```python
uni=df.select_dtypes('number')
uni=uni.drop('Depression',axis=1)
for i in uni.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[i])
```

##  Bivariate Analysis

**Bivariate Analysis** examines the relationship between two variables — for example, how one feature affects another.  
It helps identify correlations, patterns, or dependencies between features.

- **Numerical vs Numerical:** Use scatter plots or correlation heatmaps.  
- **Categorical vs Numerical:** Use boxplots or barplots.  
- **Categorical vs Categorical:** Use countplots or crosstab tables.

###  Barplot
A **barplot** shows the relationship between a categorical variable (x-axis) and a numerical variable (y-axis), displaying the average or aggregate values for each category.

###  Countplot
A **countplot** displays the count (frequency) of each category in a single categorical feature.  
It’s useful for visualizing class balance or distribution in categorical data.

```python
# bivariate analysis for the nunmerical features
target='Depression'
biv=df.select_dtypes('number')
biv.drop('Depression',axis=1,inplace=True)
for i in biv.columns:
    plt.figure(figsize=(8,4))
    sns.barplot(x=df[target], y=df[i])

```

##  GroupBy Analysis

**GroupBy Analysis** is used to split data into groups based on one or more categorical features,  
and then apply aggregation functions (like mean, sum, count, etc.) to analyze patterns within each group.

It helps compare numerical statistics across different categories — for example,  
average scores by gender, income by city, or depression level by age group.

```python
# groupby analysis
city_depression_count = df.groupby('City')['Depression'].mean().sort_values(ascending=False)
sns.barplot(x=city_depression_count.index, y=city_depression_count.values)
plt.xticks(rotation=90)
plt.title("Total Number of Depressed People per City")
plt.xlabel("City")
plt.ylabel("Depression Count (1s)")
plt.show()
```

```python
degree_depression_count = df.groupby('Degree')['Depression'].mean().sort_values(ascending=False)
sns.barplot(x=degree_depression_count.index, y=degree_depression_count.values)
plt.xticks(rotation=90)
plt.title("Total Number of Depressed People based on Degree")
plt.xlabel("Degree")
plt.ylabel("Depression Count (1s)")
plt.show()
```

##  Feature Scaling in EDA

**Feature Scaling** is the process of transforming numerical features into a similar range  
to ensure that no single feature dominates the model due to its scale.  
It is especially important for distance-based algorithms (like KNN, SVM) and gradient-based models.

### 🔹 Normalization (Min-Max Scaling)
**Min-Max Scaling** transforms features to a fixed range, usually [0, 1], using the formula:

```python
f_scal=['Age','Academic Pressure', 'CGPA', 'Study Satisfaction', 'Study Hours','Financial Stress']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler


preprocessor=ColumnTransformer(
    transformers=[
        ('minmax',MinMaxScaler(),f_scal)
    ],
    remainder='passthrough'
)
```
