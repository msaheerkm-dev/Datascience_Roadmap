```python
import nltk
```

```python
from nltk.stem import WordNetLemmatizer
wlem=WordNetLemmatizer()
```

```python
words=['running','run','going','goes','playing','history','teacher','teaching']
for word in words:
    print(word+"------>",wlem.lemmatize(word,pos='v'))
```

```python
# pos has 4 values
# n = nown 
# v = verb
# a = adjective
# r = asverb
```

### stopwords

```python
corpus="""Artificial Intelligence (AI) is a technology that allows machines to learn from data, recognize patterns, and make decisions with minimal human help. It powers many tools we use every day, such as voice assistants, recommendation systems, and smart cameras. AI can analyze huge amounts of information quickly, helping solve complex problems in fields like healthcare, finance, and education. As AI continues to grow, it promises to make our lives easier and more efficient.
"""
print(corpus)
```

```python
from nltk import sent_tokenize

documnets=sent_tokenize(corpus)
documnets
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
from nltk.tokenize import word_tokenize
words=word_tokenize(corpus)
words
```

```python
new_words=[wlem.lemmatize(word) for word in words if word not in stopwords.words('english')]
new_words
sentence = ' '.join(new_words)
print(sentence)
```
