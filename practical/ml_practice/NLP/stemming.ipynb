```python
# PorterStemmer
from nltk.stem import PorterStemmer
ps=PorterStemmer()

words=['running','run','going','goes','playing','history','teacher','teaching']
for word in words:
    print(word+"------>",ps.stem(word))
```

```python
# RegexStemmer
from nltk.stem import RegexpStemmer
rs=RegexpStemmer('ing$|s$|ed$|es$|er$',min=4)

words=['running','run','going','goes','playing','history','teacher','teaching']
for word in words:
    print(word+"------>",rs.stem(word))
```

```python
# snowballStemmer
from nltk.stem import SnowballStemmer
ss=SnowballStemmer('english')

words=['running','run','going','goes','playing','history','teacher','teaching']
for word in words:
    print(word+"------>",ss.stem(word))
```
