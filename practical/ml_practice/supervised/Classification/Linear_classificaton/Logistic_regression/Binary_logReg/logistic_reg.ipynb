```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
df = pd.read_csv('insurance_data.csv')
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
# Step 3: Visualize relationship (optional)
sns.scatterplot(x='age', y='bought_insurance', data=df)
plt.title('Age vs Bought Insurance')
plt.show()
```

```python
# Step 4: Split features and target
X = df[['age']]               # Independent variable
y = df['bought_insurance']    # Target variable (0 or 1)
```

```python
# check for class imbalance
sns.countplot(x=y,data=df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
# Step 6: Create and train model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Step 7: Predict
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
